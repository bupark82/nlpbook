SQL 및 데이터베이스 아키텍처 : 이것은 하나 이상의 과정이 될 가능성이 높습니다. 데이터베이스가 작동하는 방식, 데이터를 저장하는 방법 및 데이터를 효율적으로 읽고 쓸 수 있도록 데이터베이스를 관리하는 방법을 배웁니다. SQL(일반적으로 "sequel"이라고 발음하며 Structured Query Language를 의미함)이라는 언어를 사용하는 방법을 배웁니다. 데이터는 일반적으로 설계상 스프레드시트와 유사한 테이블에 저장되며 SQL은 이러한 데이터를 찾는 데 이상적입니다. 테이블은 서로 관련되어 있으므로 "관계형 데이터"라고 합니다. MongoDB와 같은 비관계형 데이터베이스에 대한 과정을 수강할 수도 있습니다. 비관계형 데이터베이스는 SQL을 사용하지 않으며 일반적으로 "NoSQL"("후속 없음"으로 발음)이라고 합니다.

빅 데이터 아키텍처 : Hadoop 및 Spark와 같은 이름을 가진 도구를 사용하여 "빅 데이터"로 알려지게 된 것을 관리하는 방법을 배웁니다. 즉, Google 및 Amazon과 같은 거대 기업 및 기술 회사에서 일반적으로 발견되는 수십억 개의 데이터 레코드입니다. 이러한 유형의 데이터 처리는 처리를 여러 컴퓨터에 분산시켜야 하는 경우가 많기 때문에 복잡해집니다. Hadoop과 같은 도구는 이러한 종류의 작업에 이상적입니다.

데이터 시각화 : 여기에서 하나 이상의 과정을 수강해야 할 것입니다. 데이터를 분석하고 궁극적으로 제시하기 위해 다양한 도구를 사용하게 됩니다. Python은 종종 matplotlib, numpy 및 pandas와 같은 이름을 가진 도구와 함께 이 과정에 제공됩니다.

클라우드 컴퓨팅 : 빅 데이터는 컴퓨터의 큰 네트워크에서 처리되어야 합니다. Amazon Web Services와 같은 클라우드 공급자에는 Hadoop 및 기타 "빅 데이터" 도구를 실행하는 서비스가 포함됩니다. 이러한 서비스는 필요한 소프트웨어가 설치된 여러 대의 컴퓨터도 할당합니다.
